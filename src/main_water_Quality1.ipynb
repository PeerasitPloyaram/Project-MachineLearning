```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Import CSV

```python
df = pd.read_csv('../data/waterQuality1.csv')
# https://www.kaggle.com/datasets/mssmartypants/water-quality/data
```

```python
df
```

Preprocessing Data

```python
df = df.drop(df[df['ammonia'] == "#NUM!"].index)

df['ammonia'] = df['ammonia'].astype('float')
df['is_safe'] = df['is_safe'].astype('float')
```

```python
df
```

```python
df.info()
```

Split to x and y

```python
x = np.array(df.drop(['is_safe'], axis=1))
y = np.array(df['is_safe'].copy())

from model_selection import positive_negative_check
p,s = positive_negative_check(y)
```

Change negative and positve class from 0, 1 to -1, 1

```python
y = np.where(y <= 0, -1, 1)
p, s = positive_negative_check(y)
```

Split x and y to train, test and validation 

```python
from model_selection import train_test_split

buff_x, x_test, buff_y, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
x_train, x_validate, y_train, y_validate = train_test_split(buff_x,buff_y, test_size=0.3, random_state=13)
# x_train
```

SVM Classifier

```python
from model import SVM
model1 = SVM(kernel='linear', learning_rate=0.001, max_itr=1000, debug=True, lambda_param=0.01)
model1.fit(x_train, y_train)

predict_test = model1.predict(x_test)
predict_validate = model1.predict(x_validate)

print("Validate Accuracy:",model1.score(x_validate, y_validate))
for i in range(50):
    print(predict_test[i], y_test[i])
```

```python
label = ['Positve Class: 1', 'Negative Class: -1']
myexplode = [0.2, 0]
plt.figure(figsize=(5,5))
plt.pie([p,s], labels=label,shadow=True, explode=myexplode, autopct='%1.2f%%', startangle=10)
plt.legend()
plt.show()
```
