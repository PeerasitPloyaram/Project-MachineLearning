```python
import numpy as np
import pandas as pd
from model_selection import train_test_split, positive_negative_check
from sklearn import svm
from model import SVM
```

```python
df = pd.read_csv("../archive-2/breast-cancer-wisconsin.csv")
df = df.drop(['id'], axis=1)

df = df.drop(df[df['bare_nuclei'] == '?'].index)
df['bare_nuclei'] = df['bare_nuclei'].astype(np.int64)
df
```

```python
x = np.array(df.drop(['class'], axis=1))
y = np.array(df['class'].copy())

y = np.where(y == 2, -1 , 1)

for i in x:
    print(i)
```

```python
buff_x, x_test, buff_y, y_test = train_test_split(x, y, random_state=42)
x_train, x_validate, y_train, y_validate = train_test_split(buff_x, buff_y, random_state=42)
```

```python
model1 = svm.SVC(kernel='linear',C=100)
model1.fit(x_train, y_train)
print(model1.score(x_validate, y_validate))
print(model1.score(x_test, y_test))
```

```python
model2 = SVM(kernel='linear',c=100,learning_rate=0.001, lambda_param=0.001, debug=True, max_itr=2000)
model2.fit(x_train, y_train)
predict = model2.predict(x_test)
predict_validate = model2.predict(x_validate)

print(model2.score(predict_validate, y_validate))
print(model2.score(predict, y_test))
```
